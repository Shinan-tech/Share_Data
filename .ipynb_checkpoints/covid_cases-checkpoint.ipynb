```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv

df = pd.read_csv("data/countries-aggregated.csv")
df = df.where(df['Country'] == 'Germany').dropna(how='all')
df.head()
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
ax = df.reset_index().plot(x='Date', y='Confirmed', figsize=(120, 10), marker= 'o')
ax.figure.savefig('images/confirmed_cases_germany.svg', format='svg', dpi=1000)
```

```python
ax = df.reset_index().plot(x='Date', y='Recovered', figsize=(120, 10), marker= 'o')
ax.figure.savefig('images/recovered_cases_germany.svg', format='svg', dpi=1000)
```

```python
ax = df.reset_index().plot(x='Date', y='Deaths', figsize=(120, 10), marker= 'o')
ax.figure.savefig('images/death_cases_germany.svg', format='svg', dpi=1000)
```
