```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
```

```python
url ="https://de.wikipedia.org/wiki/DAX"
```

```python
html = requests.get(url).content
index_dax = bs(html, "lxml")
print(index_dax.prettify())
```

```python
info_box = index_dax.find("table", {"id":"zusammensetzung"})
info_box
```

```python
info_box = index_dax.find("table", {"id":"zusammensetzung"})
columns = info_box.find("tbody").find_all("th")
column_names = [c.text.replace("\n", "").replace("\xa0%", "") for c in columns]
table_rows = info_box.find("tbody").find_all("tr")
l = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [re.sub(r"[\n\xa0]", "", tr.text) for tr in td]
    l.append(row)
# print(l[1:30])

df_index = pd.DataFrame(l[1:30], columns=column_names)

df_index.to_pickle("Stocklist_DAX") 

```
