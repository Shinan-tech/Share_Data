```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
pytrend = TrendReq()
pytrend.build_payload(kw_list=['Coronavirus'], timeframe='2020-01-01 2020-09-30', cat=0, geo='DE')
```

```python
df = pytrend.interest_over_time()
```

```python
df
```

```python
ax = df.reset_index().plot(x='date', y='Coronavirus', figsize=(20, 10), marker ='o')
ax.set_xlabel("Month")
ax.set_ylabel("Trend")
ax.figure.savefig('images/coronavirus_trend_over_time.png', format='png')
```

```python
related_queries = pytrend.related_queries()
list(related_queries.values())[0].keys()
top_df    = list(related_queries.values())[0]['top']
rising_df = list(related_queries.values())[0]['rising']
top_df
```

```python
ax = top_df[::-1].reset_index().plot(x='query', y='value', figsize=(20, 10), kind='bar')
ax.set_xlabel("Topic")
ax.set_ylabel("Trend")
ax.figure.savefig('images/trend_of_related_topics.png', format='png')
```

```python
region_df = pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)
region_df
```

```python
ax = region_df.reset_index().plot(x='geoName', y='Coronavirus', figsize=(20, 10), kind='bar')
ax.set_xlabel("Region")
ax.set_ylabel("Trend")
ax.figure.savefig('images/coronavirus_interest_by_region.png', format='png')
```
