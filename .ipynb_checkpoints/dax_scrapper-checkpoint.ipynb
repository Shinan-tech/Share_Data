```python
import requests
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
st_date  = datetime(2020, 1, 1)
end_date = datetime.now()
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

dax_daily_url = f'https://in.investing.com/indices/germany-30-historical-data?end_date={int(datetime.timestamp(end_date))}&interval_sec=daily&st_date={int(datetime.timestamp(st_date))}'

dax_weekly_url = f'https://in.investing.com/indices/germany-30-historical-data?end_date={int(datetime.timestamp(end_date))}&interval_sec=weekly&st_date={int(datetime.timestamp(st_date))}'

dax_monthly_url = f'https://in.investing.com/indices/germany-30-historical-data?end_date={int(datetime.timestamp(end_date))}&interval_sec=monthly&st_date={int(datetime.timestamp(st_date))}'
```

```python
result_daily = requests.get(url=dax_daily_url,headers=headers)
with open(file='files/dax_daily.html', mode='w+', encoding='utf-8') as dax_file:
    dax_file.write(result_daily.content.decode())
    dax_file.flush()
    dax_file.close()

result_weekly = requests.get(url=dax_weekly_url,headers=headers)
with open(file='files/dax_weekly.html', mode='w+', encoding='utf-8') as dax_file:
    dax_file.write(result_weekly.content.decode())
    dax_file.flush()
    dax_file.close()

result_monthly = requests.get(url=dax_monthly_url,headers=headers)
with open(file='files/dax_monthly.html', mode='w+', encoding='utf-8') as dax_file:
    dax_file.write(result_monthly.content.decode())
    dax_file.flush()
    dax_file.close()
```

```python
dax_daily_df = pd.read_html('files/dax_daily.html')[1]
dax_daily_df.to_csv('data/dax_daily.csv',index=None)

dax_weekly_df = pd.read_html('files/dax_weekly.html')[1]
dax_weekly_df.to_csv('data/dax_weekly.csv',index=None)

dax_monthly_df = pd.read_html('files/dax_monthly.html')[1]
dax_monthly_df.to_csv('data/dax_monthly.csv',index=None)
```

```python
print('DAX Daily Data')
print(dax_daily_df)
print()
print('DAX Weekly Data')
print(dax_weekly_df)
print()
print('DAX Monthly Data')
print(dax_monthly_df)
```

```python
ax_d = dax_daily_df[::-1].reset_index().plot(x='Date', y='Price', figsize=(100, 40), marker= 'o')
ax_d.set_xlabel("Date")
ax_d.set_ylabel("DAX Price")
ax_d.figure.savefig('images/DAX per day.svg', format='svg', dpi=1000)

ax_w = dax_weekly_df[::-1].reset_index().plot(x='Date', y='Price', figsize=(80, 40), marker= 'o')
ax_w.set_xlabel("Date")
ax_w.set_ylabel("DAX Price")
ax_w.figure.savefig('images/DAX per week.svg', format='svg', dpi=1000)

ax_m = dax_monthly_df[::-1].reset_index().plot(x='Date', y='Price', figsize=(60, 40), marker= 'o')
ax_m.set_xlabel("Month")
ax_m.set_ylabel("DAX Price")
ax_m.figure.savefig('images/DAX per month.svg', format='svg', dpi=1000)
```
