```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Corona_Cases/owid-covid-data.csv")
df = df.where(df['location'] == 'Germany').dropna(how='all').reset_index()
df.head()
```

```python
df_new=df[['date','new_cases','new_deaths']]
df_new
```

```python
df_new_loc = df_new.loc[1:274, ["date","new_cases", "new_deaths"]]
df_new_loc
```

```python
df_new_loc.date = pd.to_datetime(df_new_loc.date)
df_new_loc
```

```python
ax = df_new_loc.reset_index().plot(x='date', y='new_cases', figsize=(120, 10), marker= 'o')
```

```python
ax = df_new_loc.reset_index().plot(x='date', y='new_deaths', figsize=(120, 10), marker= 'o')
```
